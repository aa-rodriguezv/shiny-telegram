```python
import pandas as pd
```

```python
df_snow = pd.read_csv('snow.csv')
```

```python
df_snow.head(5)
```

```python
df_resorts = pd.read_excel('resorts.xlsx')
```

```python
df_resorts.head(5)
```

```python
df_resorts.columns
```

```python
df_resorts = df_resorts[['ID', 'Latitude', 'Longitude']]
```

```python
df_resorts.set_index('ID', inplace=True)
df_resorts.head(5)
```

```python
df_resorts.columns
```

```python
dict_resorts = df_resorts.to_dict('index')
```

```python
from geopy.distance import geodesic as GD
```

```python
def check_closest_resort(lat, lon):
    resulting_resort = 0
    resulting_distance = 30
    for key, value in dict_resorts.items():
        coord_snow = (lat, lon)
        coord_resort = (value['Latitude'], value['Longitude'])
        test_distance = GD(coord_resort, coord_snow).km
        if test_distance < 27 and test_distance < resulting_distance:
            resulting_resort = key
            resulting_distance = test_distance
    return resulting_resort
```

```python
df_snow_copy = df_snow
df_snow_copy.head(5)
```

```python
df_snow_copy['Snow'] = df_snow_copy['Snow'].map(lambda x: x/100)
df_snow_copy.head(5)
```

```python
df_snow_copy_summary = df_snow_copy[['Latitude', 'Longitude']]
df_snow_copy_summary.head(5)
```

```python
df_snow_copy_summary = df_snow_copy_summary.groupby(['Latitude', 'Longitude']).count()
df_snow_copy_summary.head(5)
```

```python
df_snow_copy_summary.reset_index(inplace=True)
```

```python
df_snow_copy_summary['ID_Resort'] = df_snow_copy_summary.apply(lambda x: check_closest_resort(x['Latitude'], x['Longitude']), axis=1)
df_snow_copy_summary.head(10)
```

```python
print('Success!')
```

```python
df_snow_added_resort = pd.merge(df_snow_copy, df_snow_copy_summary, on=['Latitude', 'Longitude'], how='left')
df_snow_added_resort.head(10)
```
